- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````
def safe_divide(a, b):
    try:
      if not isinstance(a,(int,float)) or not isinstance(b,(int,float)):
        raise TypeError("Inputs must be numeric")
      return a/b
    except ZeroDivisionError:
      return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None
try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"
print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[4, 1],
              [2, 3]], dtype=float)
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors (columns):")
print(eigenvectors)

for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]

    left_side = A @ v
    right_side = lam * v

    print(f"\nChecking eigenpair {i+1}")
    print("A @ v =", left_side)
    print("λ * v =", right_side)

    assert np.allclose(left_side, right_side)
fig, ax = plt.subplots()

for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]

    scaled_v = lam * v

    ax.quiver(0, 0,
              scaled_v[0], scaled_v[1],
              angles='xy',
              scale_units='xy',
              scale=1,
              label=f"eigenvector {i+1}")

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(0, color='black', linewidth=0.8)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Eigenvectors scaled by Eigenvalues")

ax.legend()

plt.grid(True)
plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

> *Eigenvectors are special vectors that keep the same direction after a matrix transformation.
When a matrix acts on an eigenvector, it only changes the vector’s length, not its direction.

The eigenvalue tells how much the vector is scaled:
- If the eigenvalue is greater than 1, the vector stretches.
- If it is between 0 and 1, the vector gets squished or shrinks.
- If it is negative, the vector flips direction while also scaling.

Geometrically, a matrix transformation can rotate, stretch, compress, or reflect vectors in space.
Eigenvectors show the directions that remain unchanged by the transformation, and eigenvalues measure how strongly the transformation affects those directions.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````
import numpy as np

np.random.seed(42)

X = np.random.randn(4, 3)

# computing SVD
U, S, Vt = np.linalg.svd(X, full_matrices=False)

# reconstructing original matrix
X_reconstructed = U @ np.diag(S) @ Vt

# rank-1 approximation
X_approx = S[0] * np.outer(U[:, 0], Vt[0, :])

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")

print("\nSingular values:")
print(S)

print(f"\nReconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")

print("\nRank-1 approximation:")
print(X_approx)

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"

print("\n4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

> *In Singular Value Decomposition (SVD), a matrix X is decomposed as:

X = U S Vᵀ

where:

- U contains the left singular vectors. These represent patterns or directions in the rows of the data.
- S is a diagonal matrix containing singular values, which measure the importance or strength of each component.
- Vᵀ contains the right singular vectors, which represent directions in the feature space.

In relation to PCA:
- The columns of V (or rows of Vᵀ) are the principal components.
- These vectors give the directions of maximum variance in the data.
- The singular values in S indicate how much variance is captured along each principal component.

So, the matrix that contains the directions of maximum variance is Vᵀ (or equivalently V before transposing).

### 5.3 Error Metrics

Given these predictions and actuals, **implement from scratch** (no sklearn):
- MAE, MSE, RMSE
- R² Score
- Adjusted R² (assume 2 features)

````
import numpy as np

y_true = np.array([3.0, 5.0, 2.5, 7.0, 4.5, 6.0, 1.5, 8.0])
y_pred = np.array([2.8, 5.2, 2.1, 7.5, 4.0, 6.3, 2.0, 7.8])

n, p = len(y_true), 2

# Mean Absolute Error
mae = np.mean(np.abs(y_true - y_pred))

# Mean Squared Error
mse = np.mean((y_true - y_pred) ** 2)

# Root Mean Squared Error
rmse = np.sqrt(mse)

# R² Score
ss_total = np.sum((y_true - np.mean(y_true)) ** 2)
ss_residual = np.sum((y_true - y_pred) ** 2)

r2 = 1 - (ss_residual / ss_total)

# Adjusted R²
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MAE:        {mae:.4f}")
print(f"MSE:        {mse:.4f}")
print(f"RMSE:       {rmse:.4f}")
print(f"R²:         {r2:.4f}")
print(f"Adj. R²:    {adj_r2:.4f}")
````

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````
import numpy as np
from scipy import stats
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)

# generating samples
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test against standard normal distribution
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')

print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# creating non-stationary time series
t = np.arange(200)

ts = 0.05 * t + np.random.normal(0, 1, 200)

# ADF test
adf_result = adfuller(ts)

print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# differencing once
ts_diff = np.diff(ts)

# ADF test after differencing
adf_diff = adfuller(ts_diff)

print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````
import numpy as np
import matplotlib.pyplot as plt

def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    """

    epsilon = 1e-10

    # creating bin edges from expected distribution
    breakpoints = np.linspace(0, 100, bins + 1)
    bin_edges = np.percentile(expected, breakpoints)

    # histogram counts
    expected_counts, _ = np.histogram(expected, bins=bin_edges)
    actual_counts, _ = np.histogram(actual, bins=bin_edges)

    # converting to percentages
    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    # adding epsilon to avoid division/log by zero
    expected_pct = expected_pct + epsilon
    actual_pct = actual_pct + epsilon

    # PSI calculation
    psi = np.sum(
        (actual_pct - expected_pct) *
        np.log(actual_pct / expected_pct)
    )

    return psi


np.random.seed(1)

# training distribution
train_dist = np.random.normal(50, 10, 1000)

# shifted production distribution
drift_dist = np.random.normal(65, 12, 1000)

psi_value = compute_psi(train_dist, drift_dist)

print(f"PSI: {psi_value:.4f}")

print(
    f"Shift severity: "
    f"{'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}"
)

# plotting distributions
fig, ax = plt.subplots()

ax.hist(train_dist, bins=30, alpha=0.6, density=True, label='Training Distribution')
ax.hist(drift_dist, bins=30, alpha=0.6, density=True, label='Production Distribution')

ax.legend()
ax.set_title('Distribution Shift Visualization')

plt.tight_layout()
plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**


Concept drift: The relationship between input features and the target variable changes over time, causing the model’s predictions to become less accurate.

Covariate drift: The distribution of input features changes over time, while the relationship between inputs and outputs may remain the same.

PSI < 0.1: No significant distribution shift; the model input distribution is considered stable.

PSI 0.1–0.2: Minor shift detected; monitoring is recommended.

PSI > 0.2: Major distribution shift; model performance may degrade and retraining should be considered.

Retraining trigger example: Retrain the model when PSI stays above 0.2 for multiple monitoring periods or when prediction accuracy drops below a defined threshold.

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````
p_spam = 0.30
p_free_given_spam = 0.80
p_free_given_ham = 0.05

p_ham = 1 - p_spam

# P(free) using total probability
p_free = (
    p_free_given_spam * p_spam
    + p_free_given_ham * p_ham
)

# Bayes theorem
p_spam_given_free = (
    p_free_given_spam * p_spam
) / p_free

print(f"P('free'): {p_free:.4f}")

print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")


def naive_bayes_predict(prior_spam,
                        p_word_given_spam,
                        p_word_given_ham):

    prior_ham = 1 - prior_spam

    # evidence
    p_word = (
        p_word_given_spam * prior_spam
        + p_word_given_ham * prior_ham
    )

    # posterior probability
    posterior = (
        p_word_given_spam * prior_spam
    ) / p_word

    return posterior


pred = naive_bayes_predict(0.30, 0.80, 0.05)

assert abs(pred - p_spam_given_free) < 1e-9

print(f"naive_bayes_predict: {pred:.4f}")

print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term        | Formula               | In this example |
|-------------|----------------------|-----------------|
| Prior       | P(Spam)              | Probability that an email is spam before seeing the word "free" (0.30) |
| Likelihood  | P(word \| Spam)      | Probability that the word "free" appears in a spam email (0.80) |
| Evidence    | P(word)              | Overall probability of seeing the word "free" in any email |
| Posterior   | P(Spam \| word)      | Probability that the email is spam after observing the word "free" |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(7)

# population from exponential distribution
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

# generating sample means
sample_means = np.array([
    np.random.choice(population, sample_size).mean()
    for _ in range(n_samples)
])

# population statistics
pop_mean = population.mean()
pop_std = population.std()

# CLT predicted standard deviation
clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")

print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")

print(f"CLT predicted std: {clt_std:.4f}")

# plotting
fig, ax = plt.subplots()

# histogram of sample means
ax.hist(sample_means,
        bins=40,
        density=True,
        alpha=0.7,
        label='Sample Means')

# theoretical normal distribution
x = np.linspace(
    sample_means.min(),
    sample_means.max(),
    300
)

ax.plot(
    x,
    stats.norm.pdf(x, pop_mean, clt_std),
    linewidth=2,
    label='CLT Normal Approximation'
)

ax.set_title('CLT: Distribution of Sample Means')

ax.legend()

plt.tight_layout()
plt.show()

# KS test
ks_result = stats.kstest(
    sample_means,
    'norm',
    args=(pop_mean, clt_std)
)

print(
    f"KS test p-value: {ks_result.pvalue:.4f} "
    f"→ Approximately normal? {ks_result.pvalue > 0.05}"
)
````

**CLT Reflection (6.4):**

> The Central Limit Theorem matters because it shows that the distribution of sample means becomes approximately normal as the sample size increases, even when the original population is not normally distributed. This allows statisticians and machine learning practitioners to use normal-based methods for estimation, confidence intervals, and hypothesis testing.

The CLT justifies the assumption of approximate normality used in many classical statistical tests such as t-tests, z-tests, linear regression inference, and confidence interval calculations.
