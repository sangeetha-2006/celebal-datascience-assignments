# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
import re
import json

# -------------------------------
# 🛠️ TOOL 1: Calculator
# -------------------------------

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# -------------------------------
# 🛠️ TOOL 2: Keyword Extractor
# -------------------------------

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        stop_words = {
            "the", "is", "and", "are", "this", "that",
            "from", "with", "have", "will", "your",
            "into", "about", "there", "their", "would"
        }

        words = re.findall(r"\b[a-zA-Z]+\b", text.lower())

        keywords = []
        seen = set()

        for word in words:
            if len(word) > 4 and word not in stop_words and word not in seen:
                keywords.append(word)
                seen.add(word)

        return keywords[:5]

    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# -------------------------------
# 🤖 AGENT FUNCTION
# -------------------------------

def agent(query: str):
    query_lower = query.lower()

    try:
        # Route to Calculator
        if "calculate" in query_lower:
            expression = query.replace("Calculate", "").replace("calculate", "").strip()
            result = calculator(expression)

            return {
                "type": "calculator",
                "result": result
            }

        # Route to Keyword Extractor
        elif "keywords" in query_lower:
            text = query.replace("Extract keywords from", "").replace("extract keywords from", "").strip()
            result = extract_keywords(text)

            return {
                "type": "keyword_extractor",
                "result": result
            }

        # General Response
        else:
            return {
                "type": "general",
                "result": f"You asked: {query}"
            }

    except Exception as e:
        return {
            "type": "error",
            "result": str(e)
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# -------------------------------
# 🧪 Test Cases
# -------------------------------

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````

````python
# -------------------------------
# 🎯 Interactive Mode
# -------------------------------

while True:
    user_input = input("\nEnter query (type 'exit' to stop): ")

    if user_input.lower() == "exit":
        print("Exiting...")
        break

    response = agent(user_input)
    print("Response:", json.dumps(response, indent=4))
````
