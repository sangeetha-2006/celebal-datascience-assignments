### 1.3 Exceptions

Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Inputs must be numeric")

    try:
        return a / b
    except ZeroDivisionError:
        return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None

try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"

print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[4, 1],
              [2, 3]], dtype=float)

# 1. Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# 2. Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    Av = A @ v
    lv = lam * v

    print(f"\nEigenpair {i + 1}:")
    print("A @ v =", Av)
    print("λ * v =", lv)

    assert np.allclose(Av, lv), "Eigenvector equation failed!"

# 3. Plot eigenvectors scaled by eigenvalues
fig, ax = plt.subplots()

for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    scaled_v = lam * v

    ax.quiver(
        0, 0,
        scaled_v[0], scaled_v[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label=f"λ={lam:.2f}"
    )

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Eigenvectors Scaled by Eigenvalues")
ax.axhline(0, color='k', lw=0.8)
ax.axvline(0, color='k', lw=0.8)
ax.legend()

plt.tight_layout()
plt.show()

print("4.3 passed ✓")
````

A matrix transforms vectors. Most vectors change both direction and length, but an **eigenvector** keeps the same direction (or flips direction) after the transformation.

If

Av = λv

then **v** is an eigenvector and **λ** is its eigenvalue.

- **λ > 1**: the vector is stretched.
- **0 < λ < 1**: the vector is squished (compressed).
- **λ < 0**: the vector is flipped and scaled.

So, eigenvectors are directions that the matrix leaves unchanged, and eigenvalues tell us how much the matrix stretches or squishes along those directions.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
import numpy as np

np.random.seed(42)
X = np.random.randn(4, 3)

# 2. Compute SVD
U, S, Vt = np.linalg.svd(X, full_matrices=False)

# 3. Reconstruct X
X_reconstructed = U @ np.diag(S) @ Vt

# 4. Rank-1 approximation
X_approx = S[0] * np.outer(U[:, 0], Vt[0, :])

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

In SVD,

X = UΣVᵀ

- **U** contains the left singular vectors (directions in the sample space).
- **S (Σ)** contains the singular values, which measure the importance of each component.
- **Vᵀ** contains the right singular vectors.

For a centered dataset, the **right singular vectors** (rows of Vᵀ, or columns of V) are the **principal components** in PCA.

The directions of maximum variance are therefore contained in **V** (equivalently, the rows of **Vᵀ**). The singular values indicate how much variance is captured along each principal component, with larger singular values corresponding to more explained variance.

### 5.3 Error Metrics

Given these predictions and actuals, **implement from scratch** (no sklearn):
- MAE, MSE, RMSE
- R² Score
- Adjusted R² (assume 2 features)

````python
import numpy as np

y_true = np.array([3.0, 5.0, 2.5, 7.0, 4.5, 6.0, 1.5, 8.0])
y_pred = np.array([2.8, 5.2, 2.1, 7.5, 4.0, 6.3, 2.0, 7.8])

n, p = len(y_true), 2

# MAE
mae = np.mean(np.abs(y_true - y_pred))

# MSE
mse = np.mean((y_true - y_pred) ** 2)

# RMSE
rmse = np.sqrt(mse)

# R²
ss_res = np.sum((y_true - y_pred) ** 2)
ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
r2 = 1 - (ss_res / ss_tot)

# Adjusted R²
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MAE:        {mae:.4f}")
print(f"MSE:        {mse:.4f}")
print(f"RMSE:       {rmse:.4f}")
print(f"R²:         {r2:.4f}")
print(f"Adj. R²:    {adj_r2:.4f}")
````

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from scipy import stats
from statsmodels.tsa.stattools import adfuller
import numpy as np

np.random.seed(0)

# Generate samples
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test against standard normal N(0,1)
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')

print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# Non-stationary time series: trend + noise
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

# ADF test
adf_result = adfuller(ts)

print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)
adf_diff = adfuller(ts_diff)

print(
    f"\nAfter differencing — p-value: "
    f"{adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}"
)
````

````python
import numpy as np
import matplotlib.pyplot as plt

def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    epsilon = 1e-10

    # Use expected distribution to define bins
    breakpoints = np.percentile(expected, np.linspace(0, 100, bins + 1))
    breakpoints[0] = -np.inf
    breakpoints[-1] = np.inf

    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    psi = np.sum(
        (actual_pct - expected_pct)
        * np.log((actual_pct + epsilon) / (expected_pct + epsilon))
    )

    return psi

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)

print(f"PSI: {psi_value:.4f}")
print(
    f"Shift severity: "
    f"{'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}"
)

# Plot both distributions
fig, ax = plt.subplots()

ax.hist(train_dist, bins=30, density=True, alpha=0.6, label="Training")
ax.hist(drift_dist, bins=30, density=True, alpha=0.6, label="Production")

ax.legend()
ax.set_title('Distribution Shift Visualization')

plt.tight_layout()
plt.show()
````

- **Concept drift:** The relationship between input features and the target variable changes over time, reducing model accuracy.

- **Covariate drift:** The distribution of input features changes over time, even if the relationship to the target stays the same.

- **PSI < 0.1:** No significant distribution shift; the feature is considered stable.

- **PSI 0.1–0.2:** Moderate shift; monitor the feature and model performance closely.

- **PSI > 0.2:** Major shift; investigate the cause and consider retraining the model.

- **Retraining trigger example:** PSI remains above 0.2 for a key feature or production model performance drops below an acceptable threshold.

### 6.1 Core Concepts

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).
2. Two marbles drawn **without replacement**. Compute:
   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red, blue, green = 4, 3, 3

# Single-draw probabilities
p_red = red / total
p_blue = blue / total
p_green = green / total

# Joint probability: P(first=red AND second=blue)
p_red_then_blue = (red / total) * (blue / (total - 1))

# Conditional probability: P(second=blue | first=red)
p_blue_given_red = blue / (total - 1)

# Independence check
independent = (p_blue_given_red == p_blue)

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (
    p_free_given_spam * p_spam
    + p_free_given_ham * p_ham
)

# Bayes
p_spam_given_free = (
    p_free_given_spam * p_spam
) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    prior_ham = 1 - prior_spam

    evidence = (
        p_word_given_spam * prior_spam
        + p_word_given_ham * prior_ham
    )

    posterior = (
        p_word_given_spam * prior_spam
    ) / evidence

    return posterior

pred = naive_bayes_predict(0.30, 0.80, 0.05)

assert abs(pred - p_spam_given_free) < 1e-9

print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

| Term       | Formula        | In this example                                                                    |
| ---------- | -------------- | ---------------------------------------------------------------------------------- |
| Prior      | P(Spam)        | Probability an email is spam before seeing the word "free" = **0.30**              |
| Likelihood | P(word | Spam) | Probability of seeing "free" given the email is spam = **P("free" | Spam) = 0.80** |
| Evidence   | P(word)        | Overall probability of seeing "free" in any email = **P("free") = 0.275**          |
| Posterior  | P(Spam | word) | Probability the email is spam after seeing "free" = **P(Spam | "free") ≈ 0.8727**  |


### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

# Draw samples and record their means
sample_means = np.array([
    np.random.choice(population, size=sample_size, replace=True).mean()
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std = population.std()

# CLT prediction
clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot histogram + theoretical normal overlay
fig, ax = plt.subplots()

ax.hist(sample_means, bins=40, density=True, alpha=0.7, label='Sample Means')

x = np.linspace(sample_means.min(), sample_means.max(), 300)
ax.plot(
    x,
    stats.norm.pdf(x, loc=pop_mean, scale=clt_std),
    linewidth=2,
    label='CLT Normal Approximation'
)

ax.legend()
ax.set_title('CLT: Distribution of Sample Means')

plt.tight_layout()
plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(
    f"KS test p-value: {ks_result.pvalue:.4f} → "
    f"Approximately normal? {ks_result.pvalue > 0.05}"
)
````

The Central Limit Theorem is important because it allows us to approximate the distribution of sample means as normal, even when the underlying data are not normally distributed. This makes it possible to estimate confidence intervals, perform hypothesis tests, and build many statistical models using sample data.

The CLT justifies the common assumption that sampling distributions are approximately normal, which underlies many classical statistical tests such as t-tests, z-tests, and linear regression inference.
